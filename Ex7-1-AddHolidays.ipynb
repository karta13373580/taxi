```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df_train = pd.read_csv('train_hire_stats.csv')
df_train.shape
```

```python
#holidays連假的平常日
#workdays連假的補班日
train_holidays=set(['2016-02-08', '2016-02-09', '2016-02-10', '2016-02-11', '2016-02-12','2016-02-29', '2016-04-04', '2016-04-05', '2016-6-9', '2016-6-10', 
'2016-09-15', '2016-09-16', '2016-10-10', '2017-01-02', '2017-01-27', '2017-01-30'])
train_workdays = set(['2016-06-04', '2016-09-10'])
test_holidays = set(['2017-02-01','2017-02-27', '2017-2-28'])
test_workdays = set(['2017-02-18'])
```

```python
isworkday = np.ones((len(df_train),), dtype=int)
dayinweek = np.ones((len(df_train),), dtype=int)
```

```python
from datetime import datetime

for index, row in df_train.iterrows():
    #print(row['Date'])
    if row['Date'] in train_holidays:
        #print(row['Date'])
        isworkday[index] = 0
    else:
        #dd = pd.to_datetime(row['Date']).dt.date
        dd=datetime.strptime(row['Date'], "%Y-%m-%d")
        weekday = dd.weekday()
        dayinweek[index]= weekday+1
        if weekday >=5 and row['Date'] not in train_workdays:
            isworkday[index] = 0
        if row['Date'] in train_workdays:
            print(row['Date'])
```

```python
df_train['isWorkday'] = isworkday
df_train['dayinweek'] = dayinweek
```

```python
df2_train = df_train.sort_values(['Zone_ID', 'Date', 'Hour_slot'], ascending=[1, 1, 1])
```

```python
df2_train.to_csv("trainset.csv")
```
